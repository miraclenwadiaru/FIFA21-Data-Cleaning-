### 5. Write a function to convert height of players in inches to cm on the 'Height'  column and apply the function to the column using .map()

```python
#Convert inches to cm on the height column
def func(x):
    """ to convert inches to cm"""
    if '"' in x:
        y= x.replace('"', '') # to remove the symbol '"' and replace with an empty string
        y= y.split("\'") # to split it into feet and inches
        y=int(y[0])+int(y[1])*0.0833 
        y=30.48 * y   
        return int(round(y,0))
    else:
        return int(x.replace('cm',''))
df['Height']=df.Height.map(func)
```

Observation: Now we've successfully converted the height of those players in inches to cm

### 6 Write a function to convert the weight of some players that's in kg to lbs on the 'Weight' column, and use .map() to apply the function on the weight column

```python
#converting kg to lbs on the weight column
def kg_to_lbs(x):
    if 'kg' in x:
        y= x.replace('kg', '')
        y=int(y[0])*2.205 
        return int(round(y,0))
    else:
        return int(x.replace('lbs',''))
df['Weight']=df.Weight.map(kg_to_lbs)
```

### 7 Strip '€' sign from the 'Value' column using .replace(). Write a function to convert those values in 'K' to thousands and those in 'M' to millions. Then use .apply() to effect the function on the 'Value' column

```python
#standardizing the value column
df['Value']=df['Value'].str.replace('€', '')
def conversion(value):
    if value.endswith('K'):
        return float(value[:-1])*1000
    elif value.endswith('M'):
        return float(value[:-1])*10000
    else:
        return float(value)
df['Value']=df['Value'].apply(conversion)
```

### 12. write a function to split the 'Contract' column which contains the contract details of the players into    'Contract_start_year' and 'Contract_end_year'

```python
#splitting the contract column into contract_start_year and contract_end_year
def main(k):
    start=[]
    gen=k.to_list()
    for i in gen:
        start.append(i[:7])
    return start

df['Contract_Year_Start']=main(df['Contract'])

def main1(k):
    end=[]
    gen=k.to_list()
    for i in gen:
        end.append(i[-7:])
    return end
df['Contract_Year_End']=main1(df['Contract'])
df['Contract_Year_End']=df['Contract_Year_End']
```

### 12.1 print the contract_start_year column

```python
df['Contract_Year_Start']=df['Contract_Year_Start'].str.replace('~','')
df['Contract_Year_Start']
```
